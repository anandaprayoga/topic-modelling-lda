```python
import os
import pandas as pd
import random
import numpy as np
import re
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

from gensim import corpora
from gensim.models import LdaModel
```

```python
PATH = 'D://SEMESTER 7 (2024)//PPW//Topic Modelling with LDA//'
```

```python
dataset = pd.read_csv(PATH+'health.csv')
dataset = dataset['Konten']
```

```python
random.shuffle(dataset)
```

```python
dataset = dataset[0:100]
dataset
```

```python
def load_stopwords(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        stopwords = file.read().splitlines()
    return set(stopwords)

factory = StemmerFactory()
stemmer = factory.create_stemmer()

exceptions = {
    'pengalaman': 'pengalaman',
    # Tambahkan kata-kata lainnya yang mengalami masalah serupa
}

def preprocessing(text):
    text = text.lower()
    text = re.sub(r"baca juga: .*?\n", "", text)
    text = text.split('-', 1)[-1].strip()
    to_clean = ('simak breaking news dan berita pilihan kami langsung di ponselmu. '
                'pilih saluran andalanmu akses berita kompas.com whatsapp channel : '
                'https://www.whatsapp.com/channel/0029vafpbedbpzjzrk13ho3d. '
                'pastikan kamu sudah install aplikasi whatsapp ya.')
    text = text.replace(to_clean, "").strip()
    text = re.sub("[^\w\s]", "", text)
    text = re.sub("\d+", "", text)
    text = re.sub(r"\s+", " ", text).strip()
    words = text.split()
    filtered_words = []
    for word in words:
        if word in stopwords:
            continue
        if word in exceptions:
            filtered_words.append(exceptions[word])  # Gunakan kata dari exceptions
        else:
            filtered_words.append(stemmer.stem(word))  # Lakukan stemming jika tidak ada di exceptions
    return " ".join(filtered_words)
```

```python
stopwords = load_stopwords(PATH+"stopwords.txt")
data = dataset.apply(preprocessing)
```

```python
def tokenize(text):
    return text.split()
```

```python
data_token = data.apply(tokenize)
data_token[99]
```

```python
dictionary = corpora.Dictionary(data_token)

# Dokumen ke Bag of Words
corpus = [dictionary.doc2bow(text) for text in data_token]

print("Dictionary:", dictionary)
print("Corpus:", corpus)
```

```python
lda_model = LdaModel(corpus, num_topics=10, id2word=dictionary, passes=15)

topics = lda_model.print_topics(num_words=10) 
for topic in topics:
    print(topic)
```

```python
# Mengetahui proporsi topik dalam dokumen pertama
doc = corpus[97]  # Pilih dokumen yang ingin dianalisis
topic_distribution = lda_model.get_document_topics(doc)
print("Topik dalam dokumen pertama:", topic_distribution)

# Menampilkan kata-kata dalam topik pertama
top_words_in_topic_0 = lda_model.show_topic(8, topn=10)  # Menampilkan 5 kata teratas
print("Top kata dalam topik 0:", top_words_in_topic_0)
```

```python
doc_topic_distribution = [lda_model.get_document_topics(doc) for doc in corpus]
# Membuat DataFrame proporsi topik dalam dokumen
doc_topic_df = pd.DataFrame(
    [[dict(doc).get(i, 0) for i in range(lda_model.num_topics)] for doc in doc_topic_distribution],
    columns=[f"Topik_{i}" for i in range(lda_model.num_topics)]
)

doc_topic_df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Pilih dokumen tertentu (misalnya dokumen pertama)
doc_id = 8
doc_row = doc_topic_df.iloc[doc_id]

# Visualisasi
plt.figure(figsize=(10, 6))
sns.barplot(x=doc_row.index, y=doc_row.values, palette='viridis')
plt.title(f'Distribusi Topik dalam Dokumen {doc_id}')
plt.xlabel('Topik')
plt.ylabel('Proporsi')
plt.xticks(rotation=45)
plt.show()
```

```python
# Mengambil proporsi kata dalam setiap topik
topic_word_distribution = lda_model.get_topics()

# Membuat DataFrame proporsi kata dalam topik
topic_word_df = pd.DataFrame(
    topic_word_distribution,
    columns=[dictionary[i] for i in range(len(dictionary))]
)
topic_word_df
```

```python
# Pilih topik tertentu
topic_id = 6
topic_row = topic_word_df.iloc[topic_id]

# Urutkan kata berdasarkan proporsi tertinggi
top_words = topic_row.sort_values(ascending=False).head(10)

# Visualisasi
plt.figure(figsize=(10, 6))
sns.barplot(x=top_words.values, y=top_words.index, palette='viridis')
plt.title(f'Distribusi Kata dalam Topik {topic_id}')
plt.xlabel('Proporsi')
plt.ylabel('Kata')
plt.show()
```

```python
# Visualisasi
plt.figure(figsize=(8, 8))
plt.pie(top_words.values, labels=top_words.index, autopct='%1.1f%%', colors=sns.color_palette('pastel'))
plt.title(f'Distribusi Kata dalam Topik {topic_id}')
plt.show()
```

```python
# ------- Menyimpan hasil ke Excel -------
with pd.ExcelWriter(PATH+"lda_output.xlsx") as writer:
    doc_topic_df.to_excel(writer, sheet_name="Proporsi Topik per Dokumen", index=False)
    topic_word_df.to_excel(writer, sheet_name="Proporsi Kata per Topik", index=False)
```
